```python
import pandas as pd

mydata = pd.read_csv('/content/drive/MyDrive/train.csv')

for column in mydata.columns:
    if mydata[column].dtype == 'object':
        mydata[column].fillna(mydata[column].mode()[0], inplace=True)
    else:
        mydata[column].fillna(mydata[column].mean(), inplace=True)

mydata.head()
```

```python
print(mydata.columns)
```

```python

def remove_units(text):
    if isinstance(text, str):
        numerical_values = [word for word in text.split() if word.replace('.', '', 1).isdigit()]
        if numerical_values:
            return " ".join(numerical_values)
    return text

columns_to_clean = ["Mileage", "Engine", "Power", "New_Price"]

for column in columns_to_clean:
    mydata[column] = mydata[column].apply(remove_units)

for column in columns_to_clean:
    mydata[column] = mydata[column].apply(lambda x: x.strip() if isinstance(x, str) else x)

mydata
```

```python


categorical_columns = ["Fuel_Type", "Transmission"]

for column in categorical_columns:
    one_hot = pd.get_dummies(mydata[column], prefix=column)
    mydata = pd.concat([mydata, one_hot], axis=1)
    mydata.drop(column, axis=1, inplace=True)

mydata = mydata.loc[:, ~mydata.columns.duplicated()]

mydata
```

```python

from datetime import datetime

present_year = datetime.now().year

mydata['Current_Age'] = present_year - mydata['Year']

mydata
```
